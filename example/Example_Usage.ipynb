```python
import sys
sys.path.insert(0, '/Users/alex/Documents/GarNet/src/')
from garnet import parse_motifs_file
```

```python
motifs_file = "../data/multiz46way_placental.txt"
motifs = parse_motifs_file(motifs_file)
names = motifs['motifName'].unique()
accession = motifs['accession'].unique()
```

```python
names
```

```python
import mygene
mg = mygene.MyGeneInfo()
df = mg.querymany(names.tolist(), scopes=['symbol', 'name', 'alias'], fields=["HGNC", "symbol"], species="human", as_dataframe=True, returnall=True)
df
```

```python
dup = df['dup']
missing = df['missing']
df = df['out']
df = df[["HGNC", "symbol"]].dropna()
df = df.reset_index().drop_duplicates(subset='query', keep='first').set_index('query').rename_axis(None)
df
```

```python
df2 = mg.querymany(accession.tolist(), scopes=['accession'], fields=["HGNC", "symbol"], species="human", as_dataframe=True, returnall=True)
df2
```

```python
df[df.index != df.symbol]
```

```python
motifs.head()
```

```python
motifs = motifs.merge(df, how='left', left_on='motifName', right_index=True)
del motifs['HGNC']
motifs.symbol = motifs.symbol.fillna(motifs.motifName)
motifs
```

```python
motifs[motifs.motifName != motifs.symbol]
```

```python
del motifs['motifName']
motifs = motifs.rename(columns={'symbol':'motifName'})
motifs
```

```python
motifs = motifs[["ZScore","FDR_lower","motifName","orientation","chrom","LOD","motifStrand","motifStart","realhits","cid","motifScore","NLOD","BBLS","motifEnd","medianhits","accession","FDR_upper","BLS","stdevhits"]]
```

```python
motifs.rename(index=str, columns={"motifStart":"start", "motifEnd":"stop", "motifScore":"FDR", "motifStrand":"strand", "motifName":"name"}, inplace=True)
```
